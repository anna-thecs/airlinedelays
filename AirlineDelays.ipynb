```python
import pandas as pd
billboard = pd.read_csv("https://raw.githubusercontent.com/frontiertechinstitute/datasets/refs/heads/main/Billboard%20100s%20Hits%20%20Data/billboard_cleaned.csv")

billboard.head(5)
```

```python
billboard.describe()
```

```python
billboard_duplicates = billboard[billboard.duplicated(keep=False)]
billboard_duplicates
```

```python
# counts of null values
billboard.isnull().sum()
```

```python
billboard = billboard[["artist", "year", "primary_genre", "date", "weeks_at_number_one", "non_consecutive"]]

# so when you're looking at the nulls, you're only looking at a subset
nulls = billboard[billboard.isnull().any(axis=1)]

# show ALL of the rows
pd.set_option("display.max_rows", None) # max_rows attribute is what makes this work

# show ALL of the columns
pd.set_option("display.max_columns", None)

nulls
```
